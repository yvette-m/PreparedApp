```python
import gspread
from google.colab import auth
from gspread_dataframe import get_as_dataframe
from oauth2client.client import GoogleCredentials
from google.auth import default

#authenticate and create a client to interact with Google Sheets
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
spreadsheet = gc.open('NRI_Table_CensusTracts_California')
```

```python
# Select the first sheet or specify the sheet name
worksheet = spreadsheet.get_worksheet(0)
```

```python
df = get_as_dataframe(worksheet)
```

```python
# Display only rows with NA values
df_na = df[df.isna().any(axis=1)]
print(df_na)
```

```python
# Removing some columns that were completely empty (no values, not even NaN values)
df = df.dropna(axis=1, how='all')
print(df)
```

```python
# Removing rows that contain all NaN values
df = df.dropna(axis=0, how='all')
print(df)
```

```python
# Find columns with some NaN values (not completely NaN, not completely filled)
columns_with_some_nan = df.columns[df.isna().any() & ~df.isna().all()]

# Display the list of these columns
print(columns_with_some_nan)
```

```python
# Replace NaN values in the specified columns with 0
df[columns_with_some_nan] = df[columns_with_some_nan].fillna(0)
```

```python
# Checking the null values again
df.isna().sum()
```
